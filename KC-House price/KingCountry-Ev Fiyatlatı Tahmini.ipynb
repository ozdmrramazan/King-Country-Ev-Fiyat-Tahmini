```python
# Korelasyon %20 ile %90 arası corr olan sütünlar
x=new_df[['day','month','year','Age','weekday',"month_name","sqft_living","grade","sqft_above","sqft_living15","bathrooms","view","sqft_basement","bedrooms","lat","waterfront","floors","zipcode","waterfront"]]
```

```python
x=pn.get_dummies(x,drop_first=True)
```

```python
x.shape
```

```python
y=new_df["price"]
```

```python
y.head()
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
lr=LinearRegression()
regmodel=lr.fit(x_train,y_train)
```

```python
regmodel
```

```python
y_pretest=regmodel.predict(x_test)
y_pretest
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
mean_squared_error(y_test,y_pretest)**0.5   
```

```python
r2_score(y_test,y_pretest)
```
